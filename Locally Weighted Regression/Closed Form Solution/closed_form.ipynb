#### Closed Form Solution in Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate Dataset
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

Y = Y.reshape((-1,1))
X.shape, Y.shape
```

```python
# Normalize
X = (X-X.mean())/X.std()

# Visualize
plt.scatter(X,Y)
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_[:5,:])
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def closedForm(X,Y):
    
    Y = np.mat(Y)
    firstPart = np.dot(X.T,X)
    secondPart = np.dot(X.T,Y)
    
    theta = np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta = closedForm(X_,Y)
theta
```

```python
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label="prediction")
plt.legend()
plt.show()
```
