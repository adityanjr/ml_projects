```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dfx = pd.read_csv('XTrain.csv')
dfy = pd.read_csv('YTrain.csv')
df = pd.read_csv('Xtest.csv')
dfx.head(10)
```

```python
X = dfx.values
Y = dfy.values
print(X.shape)
print(Y.shape)
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def knn(X,Y,querypoint,k=10):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(X[i],querypoint)
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    vals2=vals[:,1]
    #print(vals2)
    d=sum(vals2)
    if d>4:
        return 1
    else :
        return 0
```

```python
tests = df.values
print (tests.shape)
```

```python
k=tests.shape[0]
#print(k)
k=int(k)

ans=np.full((192,1),0)
for i in range(k):
    a=knn(X,Y,tests[i],6)
    ans[i][0]=a

ans = pd.DataFrame(ans,columns=['Outcome'])
ans.head(20)
```

```python
ans = ans.to_csv('outcome.csv',index=False)
```
