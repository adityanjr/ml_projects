```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from keras.datasets import mnist
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
plt.imshow(X_test[90], cmap='gray')
plt.show()
X_test.shape,y_test.shape
```

```python
X = X_test.reshape(-1, 28*28)
y = y_test
X.shape, y.shape
```

```python
sc = StandardScaler()
X_ = sc.fit_transform(X)
X_.shape
```

```python
plt.imshow(X_[90].reshape(28,28), cmap='gray')
plt.show()
```

### Sklearn PCA

```python
pca = PCA(n_components=2)
z_pca = pca.fit_transform(X_)
z_pca.shape, pca.explained_variance_
```

### Custom PCA

```python
covar = np.dot(X_.T,X_)
covar.shape
```

```python
U,S,V = np.linalg.svd(covar)
U.shape
```

```python
Ured = U[:,:2]
Ured.shape
```

```python
Z = np.dot(X_,Ured)
Z.shape, Z
```

### Visualize Dataset

```python
new_dataset = np.hstack((Z,y.reshape(-1,1)))
df = pd.DataFrame(new_dataset, columns=['PC1','PC2','label'])
df.head()
```

```python
fg = sns.FacetGrid(df, hue='label', height=10)
fg.map(plt.scatter, 'PC1', 'PC2')
fg.add_legend()
plt.show()
```

### PCA with 784 dimensions

```python
pca = PCA()
Z_pca = pca.fit_transform(X_)
Z_pca.shape
```

```python
cum_explained = np.cumsum(pca.explained_variance_ratio_)
plt.plot(cum_explained)
plt.show()
cum_explained
```
